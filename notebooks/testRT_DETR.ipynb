```python
# 1. Install ultralytics
!pip install ultralytics
from ultralytics import RTDETR

# 2. Load your flawless 100-epoch model
model = RTDETR('/content/drive/MyDrive/VOC_PCB/rtdetr_results/train_stable/weights/best.pt')

# 3. Official Validation (Generates final Chapter 4 graphs: Confusion Matrix, PR Curves, etc.)
print("\n📊 Running Official Validation...")
metrics = model.val(
    data='/content/drive/MyDrive/VOC_PCB/yolo_ready/dataset.yaml',
    split='val',
    project='/content/drive/MyDrive/VOC_PCB/rtdetr_results',
    name='final_validation_metrics'
)

# 4. Inference / Drawing Bounding Boxes (For the visual results in your thesis)
print("\n🖼️ Running Inference & Drawing Boxes...")
results = model.predict(
    source='/content/drive/MyDrive/VOC_PCB/yolo_ready/images/val', # Reads your images
    save=True,     # <--- THIS is the command that draws the bounding boxes!
    conf=0.5,      # Only draw boxes if the AI is 50%+ confident
    project='/content/drive/MyDrive/VOC_PCB/rtdetr_results',
    name='predicted_bounding_boxes'
)
print("\n✅ Complete! All graphs and images are saved to your Google Drive.")
```
