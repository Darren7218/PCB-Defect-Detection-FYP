```python
!pip install ultralytics
from ultralytics import RTDETR
from google.colab import runtime

model = RTDETR('rtdetr-l.pt')

print("\n🚀 Starting Stable 100-Epoch RT-DETR Training...")
results = model.train(
    data='/content/drive/MyDrive/VOC_PCB/yolo_ready/dataset.yaml',
    epochs=100,
    imgsz=640,
    batch=16,

    # --- NaN FIXES ---
    amp=False,           # Disable mixed precision — eliminates float16 overflow in AIFI attention
    lr0=0.0001,          # Lower LR (was 0.001 auto-selected, too aggressive for this dataset)
    lrf=0.01,            # Final LR ratio
    cache='disk',        # Deterministic caching — avoids RAM cache non-determinism warning
    # -----------------

    optimizer='AdamW',
    weight_decay=0.0005,
    warmup_epochs=3.0,
    patience=50,
    project='/content/drive/MyDrive/VOC_PCB/rtdetr_results',
    name='train_stable',
    save=True,
    plots=True,
)

print("\n✅ Done! Shutting down to save compute units...")
runtime.unassign()
```
